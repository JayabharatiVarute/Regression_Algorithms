```python
# Import Liabraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Import dataset

ds = pd.read_csv("Salary_Data.csv")
ds
```

```python
ds.columns
```

```python
# Divide dataset into Independent and Dependent variables

x = ds[['YearsExperience']]
y = ds[['Salary']]
```

```python
# Plot graph

plt.scatter(x,y)
plt.show()
```

```python
# Splitting dataset as Training and Testing

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
# Introduce SLR Model

from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
# Predict

result = reg.predict([[10]])
result
```

```python
result = reg.predict(x_test)
result
```

```python
y_test
```

```python
# Visualize

plt.scatter(x, y)
result = reg.predict(x)
plt.plot(x, result, color ="red")
plt.show()
```

```python
# For User Input

years = int(input("Enter year of exp. : "))
myinput = pd.DataFrame(columns =['YearsExperience'], data=[[years]])
result = reg.predict(myinput)
print("Your Salary must be close to", round(result[0,0]))
```
