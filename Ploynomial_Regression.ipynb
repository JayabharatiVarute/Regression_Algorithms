```python
# Importing Libaries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Importimg and reading dataset

ds = pd.read_csv("Position_Salaries.csv")
ds
```

```python
#splitting dataset into Independent and dependent data

X=ds[["Level"]]
y=ds[["Salary"]]
```

```python
X
```

```python
y
```

```python
#plot graph

plt.scatter(X,y)
plt.show()
```

```python
# Model fitting(Polynomial regression model fitting)

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=5)
X_ploy = poly_reg.fit_transform(X)
X_ploy
reg = LinearRegression()
reg.fit(X_ploy,y)
```

```python
# visulization 

plt.scatter(X,y)
plt.plot(X, reg.predict(poly_reg.transform(X)),color='red')
plt.show()         
```

```python
# predicting for my input

myinput=pd.DataFrame(data=[[11]], columns=["Level"])
result=reg.predict(poly_reg.transform(myinput))
result

```

```python
# Predict for User Input

level = float(input("Enter Level:"))
myinput = pd.DataFrame(data=[[level]], columns=[["Levels"]])
poly_input = poly_reg.transform(myinput)
result = reg.predict(poly_input)
print(f"Predicted result for Level {level}: {round(result[0,0])}")
```
