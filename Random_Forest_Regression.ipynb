```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Improting and reading Dataset

ds = pd.read_csv("Position_Salaries.csv")
ds
```

```python
# spltting dataset into Independent & Dependent variables 

X = ds[["Level"]]
y = ds[["Salary"]]
    
```

```python
X
```

```python
y
```

```python
# fitting Random forest Regression Model 

from sklearn.ensemble import RandomForestRegressor
reg = RandomForestRegressor(n_estimators=10,random_state=0)
reg.fit(X,y)
```

```python
# predict for myinput 

myinput = pd.DataFrame(data=[[4]], columns=["Level"])
reg.predict(myinput)
```

```python
 # Visualization of random forest regressiob tree
from sklearn.tree import plot_tree
plot_tree (reg.estimators_[0],filled=True)
plt.show()
```

```python
# predict for user input

level = float(input("Enter the level:"))

myinput = pd.DataFrame(data=[[level]], columns=["Level"])
result = reg.predict(myinput)

print(f"Predicted Salary for Level {level}:₹{result[0]}")
```
